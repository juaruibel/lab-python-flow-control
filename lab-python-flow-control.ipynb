```python
products = ["t-shirt", "mug", "hat", "book", "keychain"]

inventory = {}

for item in products:
    quantity_of_item = int(input(f'Enter the quantity of {item}: '))
    inventory[item] = quantity_of_item

print(inventory)

customer_orders = set()

# Ask the user if they want to add another product (yes/no).
# Continue the loop until the user does not want to add another product.

while True:
    user_input = input(f'Enter the name of a product from {products}')
    customer_orders.add(user_input)
    question = input(f'Do you want to add another product (yes/no): ').lower()
    if question == 'no':
        break

print(customer_orders)

total_products_ordered = len(customer_orders)
percentage_of_prodcuts_ordered = total_products_ordered/len(products)*100
order_status = (total_products_ordered, percentage_of_prodcuts_ordered)

print(
    f'Order Statistics:\n'
    f'Total Products Ordered: <{total_products_ordered}>.\n'
    f'Percentage of Products Ordered: <{percentage_of_prodcuts_ordered}>%'
)

# Instead of updating the inventory by subtracting 1 from the quantity of each product, 
# only do it for  the products that were ordered (those in "customer_orders").

for item in customer_orders:
    for product in products:
        if item == product:
            inventory[item] -= 1

print(
    f'This is the upadted inventory: \n'
    f'T-shirt: {inventory["t-shirt"]}\n'
    f'Mug: {inventory["mug"]}\n'
    f'Hat: {inventory["hat"]}\n'
    f'Book: {inventory["book"]}\n'
    f'Keychain: {inventory["keychain"]}\n'
)
```
